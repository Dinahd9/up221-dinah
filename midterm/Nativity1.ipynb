```python
import geopandas as gpd
```

```python
# read in a geojson file downloaded from census reporter
tracts=gpd.read_file('data/kernnat_acs2021.geojson')
tracts.head()
```

```python
# See all the data
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# plot data
tracts.plot(figsize=(12,10))
```

```python
# inspect dataset
tracts.info(verbose=True, show_counts=True)
```

```python
# keep geoid and geometry and subset the data
tracts = tracts[['geoid','geometry']]
tracts.head()
```

```python
# Creating FIPS variable from Census Reporter geojson using the 'geoid' variable
tracts['FIPS'] = tracts['geoid']
tracts['FIPS'].head()
```

```python
## Isolating rows to create a 'FIPS' variable without row 0 (County) and creating new 'FIPS' variable
tracts = tracts.loc[1:236].copy()
tracts
```

```python
## Remove extra characters from new 'FIPS'
tracts['FIPS'] = '0'+tracts['FIPS'].str.lstrip('14000US')
tracts['FIPS'].copy()
```

```python
# check data 
tracts.head()
```

```python
## Merging Census Tract shapefile with Nativity data for kern County
kern_nativ_tracts = tracts.merge(data_file2,on="FIPS")
## Confirming that the data has been merged
kern_nativ_tracts.tail()
```

```python
kern_nativ_tracts.info()
```

```python
# map of foeign born non citizens by census tract 
kern_nativ_tracts.plot(figsize=(12,10),
                 column='Total Population: Foreign Born: Not a Citizen',
                 legend=True, 
                 cmap='Oranges',
                 scheme='NaturalBreaks')
```

```python
# map of foeign born non citizens by census tract, equal interval legend
kern_nativ_tracts.plot(figsize=(12,10),
                 column='% Total Population: Foreign Born: Not a Citizen',
                 legend=True, 
                 cmap='Oranges',
                 scheme='equal_interval')
```

```python
import folium
```

```python
m = folium.Map(location=[34.2,-118.2],
zoom_start = 9,
tiles='CartoDB positron',
attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
geo_data=kern_nativ_tracts, # geo data
data=kern_nativ_tracts, # data
key_on='feature.properties.FIPS', # key, or merge column
columns=['FIPS', '% Total Population: Foreign Born: Not a Citizen'], # [key, value]
fill_color='YlOrRd',
line_weight=0.1,
fill_opacity=0.7,
line_opacity=0.4, # line opacity (of the border)
legend_name=('Total Population: Foreign Born: Not a Citizen by Census Tract in Kern County␣
↪(2020)').add_to(m) # name on the legend color bar
m
```
