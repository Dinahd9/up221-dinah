```python
# looking at income in kern county
# defining the columns to keep
columns_to_keep = ['Geo_FIPS',
                   'SE_A14001_001',    
                   'SE_A14001_002',    
                   'SE_A14001_003',   
                   'SE_A14001_004',
                   'SE_A14001_005',
                   'SE_A14001_006',   
                   'SE_A14001_007', 
                   'SE_A14001_008',
                   'SE_A14001_009',
                   'SE_A14001_010', 
                   'SE_A14001_011',
                   'SE_A14001_012',
                   'SE_A14001_013', 
                   'SE_A14001_014',
                   'SE_A14001_015',
                   'SE_A14001_016',
                   'SE_A14001_017']
```

```python
# add to a new data frame 
data_file2 = data_file[columns_to_keep]
```
